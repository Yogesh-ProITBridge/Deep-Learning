```python
import matplotlib.pyplot as plt

# loading the image from the disk
image = cv2.imread("/content/kohli.jpeg")

# Convert the image from BGR to RGB (matplotlib uses RGB)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```

```python
import matplotlib.pyplot as plt

# loading the image from the disk
image = cv2.imread("/content/kohli.jpeg")

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python

import matplotlib.pyplot as plt


# Apply Gaussian blur
blurred_image = cv2.GaussianBlur(image, (15, 15), 0)

plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
# Apply Gaussian blur
blurred_image = cv2.GaussianBlur(image, (15, 15), 0)

# Save the blurred image
cv2.imwrite("/content/blurred_kohli.jpeg", blurred_image)

# Display the blurred image
plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread("/content/kohli.jpeg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# image rotate in 270
(h,w) = image.shape[0:2]
center = (h/2,w/2)
angle_270 = 270
scale = 1.0
rotate = cv2.getRotationMatrix2D(center,angle_270,scale)
rotated_270 = cv2.warpAffine(image,rotate,(h,w))

plt.imshow(rotated_270)
plt.show()
```
